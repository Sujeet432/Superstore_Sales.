9.2. Create a function called remove_outliers which takes as parameter the dataframe and the column which needs to be searched for outliers. Using the 3 * IQR rule, the function should detect and remove the outliers to return the cleaned dataframe

"""
Removes outliers from the specified column of the DataFrame using the 3 * IQR rule.

Parameters:
df (pd.DataFrame): The input DataFrame.
column (str): The name of the column to check for outliers.

Returns:
pd.DataFrame: A new DataFrame with outliers removed.
"""

```python
def remove_outliers(df, column):

    # Calculate Q1 and Q3
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1

    # Define outlier bounds
    lower_bound = Q1 - 3 * IQR
    upper_bound = Q3 + 3 * IQR

    # Filter out outliers
    cleaned_df = df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]

    return cleaned_df

Final_cleaned_data = remove_outliers(Final_data, 'Sales Price')
Final_cleaned_data.shape
```

11.6.3. Yearly Growth in Sales and Profit: Use a year-over-year growth chart to compare the sales and profit growth over different years.

```python
#Extract Year from Order Date
Final_cleaned_data['Order_Year'] = pd.to_datetime(Final_cleaned_data['Order Date']).dt.year.astype(str)
Final_cleaned_data['Order_Year'].unique()
```

Why convert it to str?
If you use float (like 2015.00, 2015.25), the plot will show decimal points, which look messy for years.

int can still cause ticks to be spaced oddly: Even with int, matplotlib might treat it as continuous numeric data, and space it unevenly.

str makes them categorical labels: Strings (like '2015') are treated as labels, not numbers, so they show exactly as written without any formatting issue.

Avoids decimal confusion:

When Year is in float, you get labels like 2015.0, 2016.0 — which are technically correct but look unprofessional on a time axis.

```python
#Group data by Year and aggregate Sales & Profit
yearly_data = Final_cleaned_data.groupby('Order_Year')[['Sales Price', 'Total Profit']].sum().reset_index()
print(yearly_data)
```

pandas.DataFrame.pct_change

--> DataFrame.pct_change(periods=1, fill_method=<no_default>, limit=<no_default>, freq=None, **kwargs)

Fractional change between the current and a prior element.

Computes the fractional change from the immediately previous row by default. This is useful in comparing the fraction of change in a time series of elements.

```python
# Calculate YoY growth
yearly_data['Sales YoY Growth (%)'] = yearly_data['Sales Price'].pct_change() * 100
yearly_data['Profit YoY Growth (%)'] = yearly_data['Total Profit'].pct_change() * 100

print(yearly_data)
```

```python
# Plot YoY Growth
plt.figure(figsize=(12, 6))
plt.plot(yearly_data['Order_Year'], yearly_data['Sales YoY Growth (%)'], marker='o', label='Sales YoY Growth')
plt.plot(yearly_data['Order_Year'], yearly_data['Profit YoY Growth (%)'], marker='o', label='Profit YoY Growth', color='green')
plt.title('Year-over-Year Growth in Sales and Profit')
plt.xlabel('Year')
plt.ylabel('Growth Rate (%)')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

Both sales and profit have shown consistent year-over-year growth from 2015 to 2017.

Sales YoY growth increased from around 5.6% in 2015 to approximately 26% by 2017, indicating strong revenue momentum.

Profit YoY growth also followed a similar upward trend, rising from ~2.8% in 2015 to about 25% in 2017, reflecting improved operational efficiency or cost control.

The gap between sales and profit growth narrowed over the years, suggesting that the company became more profitable as it scaled.

There are no negative growth years, which indicates stable and healthy business performance during this period.
